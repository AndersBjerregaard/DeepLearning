# Imports

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
from math import sqrt
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
```

# Read Data

```python
df = pd.read_csv('datasets/TSLA.csv')
df.info(), df.isnull().sum()
```

# Data Preparation

## Data Wrangling

```python
df.High.fillna(df.High.mean(),inplace=True)
df.Low.fillna(df.Low.mean(),inplace=True)
```

```python
df.isnull().sum()
```

## New Data Column

```python
df['Average'] = df[['Close', 'High']].mean(axis=1)
df.info(), df['Average'][:10]
```

## Visualize data to work with

```python
df.plot('Date', 'Average', figsize=(18,8))
```

## Scale Data

```python
scaler = MinMaxScaler(feature_range=(-1, 1))
# Reshape because of single feature array
df['Average'] = scaler.fit_transform(df['Average'].values.reshape(-1, 1))
df['Average']
```
