```python
# From the matplotlib package, import the pyplot module
from matplotlib import pyplot as plt
```

```python
# Simple example of how to use pyplot
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]
# Plot the x elements along the x-axis, and y elements along the y-axis
plt.plot(x, y)
# Plot the x elements with the z elements
plt.plot(x, z)
# Add title and labels to the graph and its axis'
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y and z")
# Add some explanation to the graphs
plt.legend(["this is y", "this is z"])
# Show the graph
plt.show()
```

```python
# Import some real data, in this case the country data from the countries.csv file
data = pd.read_csv('countries.csv')
```

```python
# Compare the population growth in the US and China
# To do this, the logical thing to do is to isolate the data in the US and China

# the "data.country == 'United States'" expression returns a Series of false / true.
# Meaning that if a row is 'United States' it returns true, and if it isn't it returns false.
# the above expression is wrapped in the brackets of "data[]", it takes only the collection that has the values "true"
us = data[data.country == 'United States']
```

```python
# Showcase the US data
us
```

```python
china = data[data.country == 'China']
```

```python
china
```

```python
# Plot the US population growth
# Because the numbers are so big. Divide the population values with 10^6 to show the values in millions.
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.title("Population difference between US and China")
plt.xlabel("Year")
plt.ylabel("Population in millions")
plt.legend(["US", "China"])
plt.show()
```

```python
# Find the percentage growth of the first year of the set (year 1952)
```

```python
# Divide all of the series' values to the value of the first year, multiplied by 100 to show as percentage amounts
us.population / us.population.iloc[0] * 100
# This should show the values of the population Series in relation to the value at the '0' index of the collection
```

```python
plt.plot(us.year, us.population / us.population.iloc[0] * 100 - 100)
plt.plot(china.year, china.population / china.population.iloc[0] * 100 - 100)
plt.title("Population growth difference between US and China")
plt.xlabel("Year")
plt.ylabel("Population growth in % in porportion to year 1952")
plt.legend(["US", "China"])
plt.show()
```
